## Librerias

```python
import speech_recognition as sr
from gtts import gTTS 
from playsound import playsound
from datetime import date
import calendar
import time
import pandas as pd
import random as rd 
import os
import glob
```

## Definicion de funciones, diccionarios y listas

```python
# Para simular la lista de contactos que tiene una persona.
contactos = ["fernando", "maria", "esteban", "alex"]

# Puesto que hay una funcion que nos devuelve el nombre del mes de determinada fecha pero
# en ingles, se creo este diccionario para saber el nombre del mes de determinada fecha
# pero en español, apartir del numero de un mes.
nombreMeses = {
    "01":'Enero',
    "02":'Febrero',
    "03":'Marzo',
    "04":'Abril',
    "05":'Mayo',
    "06":'Junio',
    "07":'Julio',
    "08":'Agosto',
    "09":'Septiembre',
    "10":'Octubre',
    "11":'Noviembre',
    "12":'Diciembre'
}

reverseNombreMeses = {
    'Enero':"1",
    'Febrero':"2",
    'Marzo':"3",
    'Abril':"4",
    'Mayo':"5",
    'Junio':"6",
    'Julio':"7",
    'Agosto':"8",
    'Septiembre':"9",
    'Octubre':"10",
    'Noviembre':"11",
    'Diciembre':"12"
}

# Similarmente al diccionario: "nombreMeses"
nombreDias = {
    "1":'Lunes',
    "2":'Martes',
    "3":'Miercoles',
    "4":'Jueves',
    "5":'Viernes',
    "6":'Sabado',
    "7":'Domingo',
}

# Funcion para quitar todos los acentos que pueda tener un string.
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

# Funcion para crear un audio respuesta a partir de un texto recibido.
def generarRespuesta(text1):
    language = 'es'
    speech = gTTS(text = text1, lang = language, slow = False)
    text2 = "Resp "+ text1 + ".mp3" 
    speech.save(text2)
    return text2

# Funcio para escuchar y determinar con exactitud que dijo el usuario.
def escuchar():
    r = sr.Recognizer() 
    
    with sr.Microphone() as source:
        print('Escuchando...')
        audio = r.listen(source)

        text = ' '
        try:
                text = r.recognize_google(audio, language = 'es-MX')
                print("Tu dijiste: ")
                print(text)
        except:
                print('Lo lamento, no pude escucharte.')
        
        return text
    
# Funcion para dar un mensaje aleatorio con respecto al clima a apartir de un numero recibido.
def determinarClima():
    temp = rd.randint(10,40) #Se comenzo en 10 porque era primavera (casi verano)
    
    if(temp >= 30):
        clima = "Esta haciendo mucho calor"
    elif(temp < 30 and temp >= 15):
        clima = "El clima esta bien"
    elif(temp < 15 and temp >= 1):
        clima = "Esta haciendo frio"
    else:
        clima = "Esta haciendo mucho frio"
    
    return clima

# Funcion para retornar la hora en la que se ejecute esta funcion.
def determinarHora():
    tiempo = time.ctime()
    tiempo = tiempo.split()
    return tiempo[3]
    
# Funcion para dar mensajes de despedida aleatorios
def mensajeDespedida():
    numero = rd.randint(0,40) #Se comenzo en 10 porque era primavera (casi verano)
    
    if(numero >= 30):
        clima = 'Cualquier cosa aca andamos'
    elif(numero < 30 and numero >= 15):
        clima = 'Hasta pronto'
    elif(numero < 15 and numero >= 1):
        clima = "Quedo a tus ordenes"
    else:
        numero = "Hasta la proxima"
    
    return clima

# Como al repetir todo el programa de manera infinita, esté entra 
# en conflicto si un audio respuesta ya existe, de manera que 
# se borran todos los audio respuesta.
def eliminarRespuestas():
    py_files = glob.glob('*.mp3')

    for py_file in py_files:
        try:
            os.remove(py_file)
        except OSError as e:
            print(f"Error:{ e.strerror}")

```

## Codigo

```python
text = ' '
banderaTiempo = 0
mensajeAEnviar = None
eliminarRespuestas()

# Mientras el asistente no escuche su nombre, se queda a la espera de ello. Al escuchar
# su nombre, genera una respuesta diciendo: que necesita.
while( (text.lower()) != 'celeste' ):
    text = escuchar()

respuesta = generarRespuesta('En que puedo servirle amo')
playsound(respuesta)
time.sleep(600/1000)

while(1):
    # escucha lo que dice el usuario, esto se pasa a minusculas, despues se le quita los acentos
    # y finialmente se separa por palabras.
    listWords = normalize(escuchar().lower()).split(" ")
    
    
    # MENSAJE DE TEXTO
    if ('redactar' in listWords and 'mensaje' in listWords) or ('enviar' in listWords and 'mensaje' in listWords) or ('mandar' in listWords and 'mensaje' in listWords):    
        respuesta = generarRespuesta('Listo para escribir')
        playsound(respuesta)
        time.sleep(600/1000)

        mensajeAEnviar = escuchar()

        respuesta = generarRespuesta('A quien sera enviado el mensaje')
        playsound(respuesta)
        time.sleep(600/1000)

        text = escuchar().lower().split(" ")
        posContacto = text.index("a")
        contacto = normalize(text[posContacto + 1])

        # Si la persona a la cual se le quiere enviar el mensaje se encuentra entre los contactos, 
        # se enviara el mensaje, en caso contrario se avisara que no se encontro a la persona.
        if contacto in contactos:
            respuesta = generarRespuesta('Mensaje enviado')
            playsound(respuesta)
            time.sleep(600/1000)
        else:
            mensajeAEnviar = None
            respuesta = generarRespuesta('Esa persona no se encuentra en su lista de contactos. Mensaje no enviado')
            playsound(respuesta)
            time.sleep(600/1000)


    # Para lee el mensaje enviado, si es que este fue enviado
    elif ('leer' in listWords and 'mensaje' in listWords) or ('leerme' in listWords and 'mensaje' in listWords):
        if isinstance(mensajeAEnviar, str):
            respuesta = generarRespuesta("Usted dijo " + normalize(mensajeAEnviar) + " y fue enviado a " + contacto)
            playsound(respuesta)
            time.sleep(600/1000)
        else:
            respuesta = generarRespuesta('No ha enviado mensajes recientemente')
            playsound(respuesta)
            time.sleep(600/1000)
              
            
    # FECHA
    elif ('como' in listWords and 'andamos' in listWords):
        fecha = date.today()
        partes = str(fecha)
        partes = partes.split("-") 

        respuesta = generarRespuesta("Hoy es " + partes[2] + " de " + nombreMeses[str(partes[1])] + " del anio "+ partes[0])
        playsound(respuesta)
        time.sleep(600/1000)

        listWords = normalize(escuchar().lower()).split(" ")
        
        # Para saber el nombre del dia de la semana, del dia actual.
        if ('dia' in listWords and 'semana' in listWords) or ('dia' in listWords and 'hoy' in listWords):
            respuesta = generarRespuesta("Hoy es " + nombreDias[str(fecha.weekday()+1)])
            playsound(respuesta)
            time.sleep(600/1000)
        
        # Para saber el nombre del dia de la semana, de un dia futuro pero del año actual.
        elif ('dia' in listWords and 'cae' in listWords):
            entrada = listWords
            posCae = listWords.index("cae")

            dia = str(entrada[posCae + 1])
            mes = reverseNombreMeses[ str(entrada[posCae + 3]).capitalize() ]

            temp = pd.Timestamp("2022-" + mes + "-"+ dia)
            respuesta = generarRespuesta("Ese dia cae "+ nombreDias[str(temp.dayofweek + 1)])
            playsound(respuesta)
            time.sleep(600/1000)
            
        # Para saber el nombre del dia de la semana, de un dia pasado pero del año actual.
        elif ('dia' in listWords and 'fue' in listWords):
            entrada = listWords
            posFue = listWords.index("fue")
            
            dia = str(entrada[posFue + 1])
            mes = reverseNombreMeses[ str(entrada[posFue + 3]).capitalize() ]
            
            temp = pd.Timestamp("2022-" + mes + "-"+ dia)
            respuesta = generarRespuesta("Ese dia fue "+ nombreDias[str(temp.dayofweek + 1)])
            playsound(respuesta)
            time.sleep(600/1000)
            
                
    # CALCULADORA
    elif ('calculadora' in listWords) or ('operaciones' in listWords):
        respuesta = generarRespuesta("Que operacion desea realizar")
        playsound(respuesta)
        time.sleep(600/1000)

        listWords = normalize(escuchar().lower()).split(" ")
        operacion = listWords
        posNum = listWords.index("es")
        num1 = int(operacion[ posNum + 1 ])
        num2 = int(operacion[ posNum + 3 ])
        
        if ('mas' in listWords) or ('+' in listWords):
            respuesta = generarRespuesta("El resultado es " + str(num1 + num2))
            playsound(respuesta)
            time.sleep(600/1000)

        elif ('menos' in listWords) or ('-' in listWords):
            respuesta = generarRespuesta("El resultado es " + str(num1 - num2))
            playsound(respuesta)
            time.sleep(600/1000)

        elif ('entre' in listWords) or ('/' in listWords):
            respuesta = generarRespuesta("El resultado es " + str(num1 / num2))
            playsound(respuesta)
            time.sleep(600/1000)

        elif ('por' in listWords) or ('*' in listWords):
            respuesta = generarRespuesta("El resultado es " + str(num1 * num2))
            playsound(respuesta)
            time.sleep(600/1000)
             
                    
    # ENCENDER APARATOS   
    elif ('enciende' in listWords) or ('prende' in listWords) or ('encender' in listWords) or ('prender' in listWords):
        respuesta = generarRespuesta("Encendido, amo")
        playsound(respuesta)
        time.sleep(600/1000)
        
        
    # CLIMA   
    elif ('clima' in listWords and 'ahorita' in listWords) or ('clima' in listWords and 'este momento' in listWords):
        respuesta = generarRespuesta(determinarClima())
        playsound(respuesta)
        time.sleep(600/1000)

        listWords = normalize(escuchar().lower())
        # si el usuario pregunta por la probabilidad de lluvia, se le contestara con un porcenjaje aleatorio
        if ('probabilidad' in listWords and 'lluvia' in listWords):
            respuesta = generarRespuesta("Hay una probabilidad del "+ str(rd.randint(20,90)) + " por ciento")
            playsound(respuesta)
            time.sleep(600/1000)
            
            
    # HORA Y ALARMAS
    elif ('hora' in listWords):
        tiempo = determinarHora().split(":")

        hora = tiempo[0]
        minuto = tiempo[1]
        segundo = tiempo[2]

        print("Son las " + hora + " horas con " + minuto + " minutos y " + segundo + " segundos")
        respuesta = generarRespuesta("Son las " + hora + " horas con " + minuto + " minutos y " + segundo + " segundos")
        playsound(respuesta)
        time.sleep(600/1000)

        respuesta = generarRespuesta("Acaso Necesitas colocar una alarma")
        playsound(respuesta)
        time.sleep(600/1000)
        listWords = normalize(escuchar().lower()) 
        # Alarma. Si el usuario quiere saber cuando ha pasado determinado tiempo.
        # Para tiempos singulares pe: 'Avisame cuando haya pasado 1 minuto' y cualquier otro tipo de tiempo.
        if ('hayan' in listWords and 'pasado' in listWords) or ('haya' in listWords and 'pasado' in listWords):
            banderaTiempo = 1
            respuesta = generarRespuesta('Por supuesto que si')
            playsound(respuesta)
            time.sleep(600/1000)

            oracion = listWords.split(" ")
            posTiempo = oracion.index("pasado")
            cantidad = int(oracion[posTiempo+1])
            unitiempo = oracion[posTiempo+2]

            #print(" " + str(cantidad) + ", "+ unitiempo )

            
    if banderaTiempo:
    #while(1):    
        tiempo = determinarHora().split(":")
        horaPost = tiempo[0]
        minutoPost = tiempo[1]
        segundoPost = tiempo[2]

        print( "Son las " + horaPost + " horas con " + minutoPost + " minutos y " + segundoPost + " segundos" )

        if( unitiempo == 'segundo' or unitiempo == 'segundos' ):
            if( hora == horaPost and minuto == minutoPost and str(int(segundo)+cantidad) == segundoPost ):
                respuesta = generarRespuesta('Es hora. Ya han pasado ' + str(cantidad) + " " + unitiempo)
                playsound(respuesta)
                time.sleep(600/1000)
                banderaTiempo = 0
                #break

        elif( unitiempo == 'minuto' or unitiempo == 'minutos' ):
            if( hora == horaPost and str(int(minuto)+cantidad) == minutoPost and segundo == segundoPost ):
                respuesta = generarRespuesta('Es hora. Ya han pasado ' + str(cantidad) + " " + unitiempo)
                playsound(respuesta)
                time.sleep(600/1000)
                banderaTiempo = 0
                #break

        elif( unitiempo == 'hora' or unitiempo == 'horas' ):
            if( str(int(hora)+cantidad) == horaPost and  minuto == minutoPost and segundo == segundoPost ):
                respuesta = generarRespuesta('Es hora. Ya han pasado ' + str(cantidad) + " " + unitiempo)
                playsound(respuesta)
                time.sleep(600/1000)
                banderaTiempo = 0
                #break

    
    # PREGUNTAR SI QUIERE ALGO MAS
    respuesta = generarRespuesta('Necesitas algo mas')
    playsound(respuesta)
    time.sleep(600/1000)
    
    listWords = normalize(escuchar().lower()).split(" ")
    if ('si' in listWords):
        respuesta = generarRespuesta('Te escucho')
        playsound(respuesta)
        time.sleep(600/1000)
        eliminarRespuestas()
    elif ('no' in listWords) or ('seria' in listWords and 'todo' in listWords):
        respuesta = generarRespuesta(mensajeDespedida())
        playsound(respuesta)
        time.sleep(600/1000)
        break
        
eliminarRespuestas()
```
